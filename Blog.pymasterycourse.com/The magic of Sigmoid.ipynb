## AND Gate to an example of activation functions:

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the truth table
data = {
    'Input A': [0, 0, 1, 1],
    'Input B': [0, 1, 0, 1],
    'Output (A AND B)': [0, 0, 0, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 2))

# Hide axes
ax.axis('tight')
ax.axis('off')

# Create the table
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.5, 1.5)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the inputs
x = np.array([0, 0, 1, 1])
y = np.array([0, 1, 0, 1])

# Compute the AND operation
z = x & y

# Create the 2D plot
plt.figure(figsize=(6, 6))

# Plot the inputs with different colors based on the output
for i in range(len(z)):
    if z[i] == 1:
        plt.plot(x[i], y[i], 'go', markersize=20, label=f'({x[i]}, {y[i]}) -> {z[i]}' if i == 3 else "")
    else:
        plt.plot(x[i], y[i], 'ro', markersize=20, label=f'({x[i]}, {y[i]}) -> {z[i]}' if i == 0 else "")

# Set labels and title
plt.xlabel('Input A')
plt.ylabel('Input B')
plt.title('2D Presentation of AND Gate')

# Set axis limits and grid
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid(True)

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

## The Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate values for plotting
x_values = np.linspace(-10, 10, 400)
y_values = sigmoid(x_values)
y_derivative = sigmoid_derivative(x_values)

# Plot the sigmoid function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Sigmoid Function')
plt.plot(x_values, y_derivative, label='Derivative of Sigmoid', linestyle='--')

# Add labels and title
plt.xlabel('x')
plt.ylabel('Sigmoid(x) / Sigmoid\'(x)')
plt.title('Sigmoid Function and its Derivative')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

## Hyperbolic Tangent Function and Its Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Define the derivative of tanh function
def tanh_derivative(x):
    return 1.0 - np.tanh(x)**2

# Generate data
x = np.linspace(-5, 5, 400)
y = tanh(x)

# Plot the tanh function
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='tanh(x)', color='b')

# Plot the derivative of the tanh function
y_derivative = tanh_derivative(x)
plt.plot(x, y_derivative, '--', label="tanh'(x)", color='r')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbolic Tangent Function and Its Derivative')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
